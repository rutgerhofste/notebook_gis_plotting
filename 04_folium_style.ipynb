```
import folium
import branca
import geojson
import pandas as pd
import geopandas as gpd
```

```
# load a sample geodataframe
gdf = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
gdf_temp = gdf.copy()
```

## Simple Style

```
# Create a column specifying the style
gdf_temp["style"] = gdf_temp.apply(lambda x: {'fillColor': '#ff0000', 'weight': 2, 'color': 'black'},axis=1)
```

```
features = folium.GeoJson(gdf_temp)
```

```
m = folium.Map(location = [40.7, -74],
               zoom_start=10,
               tiles='cartodbpositron')
features.add_to(m)
m
```

## Rule based Style

```
gdf_temp = gdf.copy()
max_color = gdf["pop_est"].max()
```

```
colormap = branca.colormap.LinearColormap(
    ['green', 'yellow', 'red'],
    vmin=3, vmax=max_color
)
```

```
# Create a column specifying the style
gdf_temp["style"] = gdf_temp.apply(lambda x: {'fillColor': colormap(x["pop_est"]), 'weight': 2, 'color': 'black'},axis=1)
```

```
features2 = folium.GeoJson(gdf_temp)
```

```
m2 = folium.Map(location = [40.7, -74],
               zoom_start=10,
               tiles='cartodbpositron')
features2.add_to(m2)
m2
```
